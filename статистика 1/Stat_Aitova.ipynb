```python
q1 = df['Pulse'].quantile(0.25)
q3 = df['Pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['Pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Pulse')
remove_outliers.head()
```

```python
no_rect = df.loc[(df['RectalTemperature'].isnull())]
no_rect.describe()
```

```python
for i in no_rect.columns:
    print(f'мода {i} - {df[i].mode()[0]}')
```

```python
# показатели здоровья оптимальны, если рассчитаны по моде
# дополним пропущенные значения модой
```

```python
df['RectalTemperature'].fillna(df['RectalTemperature'].mode()[0], inplace=True)
df['RectalTemperature'].fillna(df['RectalTemperature'].mode()[0]).isna().sum()
```

```python
# т.к. количество пропущенных значений пульса незначительно, а полезность частоты дыхания малозначительна из за больших колебаний, целесооборазно заполнить недостающие значения медианой
```

```python
df['Pulse'].fillna(df['Pulse'].median(), inplace=True)
df['Pulse'].fillna(df['Pulse'].median()).isna().sum()
```

```python
df['RespiratoryRate'].fillna(df['RespiratoryRate'].median(), inplace=True)
df['RespiratoryRate'].fillna(df['RespiratoryRate'].median()).isna().sum()
```

```python
# оценим те случаи, в которых пропущены значения состояния слизистой
```
