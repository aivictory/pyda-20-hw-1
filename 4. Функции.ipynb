```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2', '5455 028765'],
    '2': ['10006', '5400 028765', '5455 002299'],
    '3': []
}


def find_owner(number):
    res = 'Документ не найден в базе'
    for i in documents:
        if i['number'] == number:
            res = 'Владелец документа: ' + i['name']
    return res


def shelf(number):
    res = 'Документ не найден в базе'
    for i in directories:
        if number in directories[i]:
            res = 'Документ хранится на полке: ' + i
    return res


# Пользователь по команде “p” может узнать владельца документа по его номеру;
def people_document(arg_number):
    for doc_number in documents:
        if doc_number["number"] == arg_number:
            print(' Владелец документа: ', doc_number["name"])
            break
    else:
        print(' Такого документа - нет в архиве.')


def list_documents():
    for document in documents:
        for i in directories:
            if document['number'] in directories[i]:
                shelf_number = i
        print(f"№ {document['number']}, тип {document['type']}, владелец {document['name']}, "
              f"полка хранения {shelf_number}")


# Пользователь по команде “as” может добавить новую полку;
def add_shelf(number):
    if number in directories:
        print('Такая полка уже существует. Текущий перечень полок: ' + ', '.join(directories.keys()))
    else:
        directories[number] = []
        print('Полка добавлена. Текущий перечень полок: as' + ', '.join(directories.keys()))


# Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
def rem_shelf(number):
    if number in directories:
        if len(directories[number]) > 0:
            print('На полке есть документа, удалите их перед удалением полки. '
                  'Текущий перечень полок: ' + ', '.join(directories.keys()))
        elif len(directories[number]) == 0:
            del directories[number]
            print('Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys()))
    else:
        print('Такой полки не существует. Текущий перечень полок: ' + ', '.join(directories.keys()))


if __name__ == '__main__':
    commands = input("Введите номер команды: ")
    if commands == 'p':
        number = input("Введите номер документа: ")
        print(find_owner(number))
    if commands == 's':
        number = input("Введите номер документа: ")
        print(shelf(number))
    if commands == 'l':
        list_documents()
    if commands == 'as':
        number = input("Введите номер полки: ")
        add_shelf(number)
    if commands == 'ds':
        number = input("Введите номер полки: ")
        rem_shelf(number)
```
