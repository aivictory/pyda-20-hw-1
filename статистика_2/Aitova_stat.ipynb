```python
df = df2[['budget', 'popularity', 'release_date', 'runtime', 'vote_average', 'revenue', 'title', 'genres']]
df.head()
```

```python
# График 1
# Изменение средней длительности фильма в разные годы
```

```python
df['year'] = pd.to_datetime(df['release_date']).dt.year
df.head()
```

```python
df_1 = df.groupby('year').mean().reset_index()
df_1.head(10)
```

```python
plt.plot(df_1['year'], df_1['runtime'])
plt.title("Изменение средней длительности фильмов")
plt.xlabel("Годы")
plt.ylabel("Среднее время фильма")
plt.show()
```

```python
# Из данного графика можно понять, что изначально создатели фильмов не очень-то понимали, какой именно длительности должен быть фильм, но с годами это понимание к ним пришло.
```

```python
# График 2
# Количество выпускаемых фильмов по годам
```

```python
sns.histplot(df.year)
plt.title("Рост выпускаемых фильмов по годам")
plt.xlabel("Годы")
plt.ylabel("Количество выпускаемых фильмов")
plt.show()
```

```python
# Их этого графика мы можем сделать выводы, что производство кратно большегоо количества фильмов началось после 80-х годов и достигло пика в 10-х годах XXI века.
```

```python
# График 3
# Топ 15 фильмов
```

```python
from plotly.offline import init_notebook_mode, iplot

trace0 = go.Scatter(
    x=df['year'],
    y=df['budget'],
    name='Budget'
)

fig = go.Figure(data=trace0)

iplot(fig)
```

```python
df_2 = df[['title', 'vote_average']].sort_values('vote_average', ascending=False).head(15)
plt.barh(df_2['title'], df_2['vote_average'])
plt.title("Топ 15 фильмов всех времён")
plt.xlabel("Рейтинг")
plt.ylabel("Фильм")
plt.show()
```

```python
# В данном графике отображаются некоторые фильмы, средние оценки которых на других площадках сильно ниже, но большинство – общепризнанные хиты. Таким образом, делаем вывод, что на разных площадках популярны в основном одни и те же фильмы.
```
