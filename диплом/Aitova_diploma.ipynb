```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = 18,6
```

```python
#вынесем из данных отдельные цифры, для которых применим количественный анализ данных
```

```python
df1 = pd.read_csv('epnp.csv')
df1.head()
```

```python
df2 = pd.read_csv('epwh.csv')
df2.head()
```

```python
# сделаем попарный анализ данных из обоих датасетов
```

```python
df3 = df1.append(df2)
sns.pairplot(df3, hue="Player")
plt.show()
```

```python
# видим интересующие нас пары метрик
```

```python
sns.histplot(data=df3, x="Day", y="Session", hue="Player")
plt.show()
```

```python
sns.histplot(data=df3, x="Day", y="Player Level", hue="Player")
plt.show()
```

```python
sns.histplot(data=df3, x="Day", y="Stronghold Level", hue="Player")
plt.show()
```

```python
sns.histplot(data=df3, x="Session", y="Player Level", hue="Player")
plt.show()
```

```python
sns.histplot(data=df3, x="Session", y="Stronghold Level", hue="Player")
plt.show()
```

```python
sns.histplot(data=df3, x="Player Level", y="Stronghold Level", hue="Player")
plt.show()
```
